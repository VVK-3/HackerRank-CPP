```
import networkx as nx
import matplotlib.pyplot as plt
```

```
G = nx.read_edgelist("/content/p2p-Gnutella05.txt",create_using=nx.DiGraph(), nodetype = int)
```

```
print(nx.info(G))
```

**1.Degree Distribution**

```
degrees=nx.degree(G)
all_degrees=list(dict(degrees).values())
#print all_degrees
unique_degrees=list(set(all_degrees))
#print unique_degrees
count_degrees = []
for i in unique_degrees:
  count=all_degrees.count(i)
  count_degrees.append(count)
plt.xlabel("Degree")
plt.ylabel("# nodes")
plt.title("Degree distribution of ")
plt.plot(unique_degrees,count_degrees)
plt.show()
```

**3. Geodesic path length**

```
E=nx.read_edgelist("/content/p2p-Gnutella05.txt")
print("Example shortest path: ")
print(nx.shortest_path_length(E,source="1",target="658"))
E = nx.path_graph("/content/p2p-Gnutella05.txt")
print("Avg. shortest path length: ")
print(nx.average_shortest_path_length(E))
```

**4. Clustering coefficient & average clustering coefficient**

```
E = nx.read_edgelist("/content/p2p-Gnutella05.txt")
print("Clustering coefficient: ")
for i in nx.clustering(E).items():
  print (i)
print ("Avg clustering coefficient: ")
print (nx.average_clustering(E))
```

**5. Strongly connected components**

```
g = nx.read_edgelist("/content/p2p-Gnutella05.txt",create_using=nx.DiGraph(), nodetype=int)
print("Strongly connected componets: ")
c_list = sorted(nx.strongly_connected_components(g),key=len, reverse=True)
print(c_list)
for c in c_list:
  print(len(c))
```

**6.SCC properties**

```
G = nx.read_edgelist("/content/p2p-Gnutella05.txt",create_using=nx.DiGraph(), nodetype=int)
is_scc = nx.is_strongly_connected(G)
print("Is the network strongly connected: ")
print(is_scc)
print("Number of strongly connected components: ")
print (nx.number_strongly_connected_components(G))
print("Largest strongly connected component: ")
print(max(nx.strongly_connected_components(G), key=len))
```

**7. Sparseness**

```
E = nx.read_edgelist("/content/p2p-Gnutella05.txt")
d = nx.density(E)
print("Density: ")
print(d)
print("Sparseness(taken as 1/density): ")
print(1/d)
```
